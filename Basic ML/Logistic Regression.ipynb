```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Spiders_1.csv')
X = data['Grainsize']
Y = data['Spiders']
x_mean = np.mean(X)
y_mean = np.mean(Y)
```

```python
b1 = np.sum(np.dot(X-x_mean,Y-y_mean))/np.sum((X-x_mean)**2)
b0 = y_mean - b1*x_mean
y_linear = (b0+b1*X)
y_log = 1/(1+np.exp(-y_linear))
max_x = max(X)+100
min_x = min(X)-100
plt.plot(X,y_linear , '#58b970', label = 'Regression line')
#plt.plot(x_log,1/(1+np.exp((b0+b1*x_log))),  '#58b970', label = 'logisitc line')
plt.scatter(X, Y, color = '#ef5423', label = 'Scatter Plot')
#here we have not set the threshold to classify it as 1 or 0
plt.scatter(X, y_log, color = "blue", label = "Predicted")
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_squared_error

X_1 = X.reshape(-1,1)
reg = LogisticRegression()
reg = reg.fit(X_1, Y)
Y_1 = reg.predict(X_1)
plt.scatter(X_1, Y, color = "green", label = "Original")
plt.scatter(X_1, Y_1, color = "orange", label = "Predicted")
plt.legend()
plt.show()
```
